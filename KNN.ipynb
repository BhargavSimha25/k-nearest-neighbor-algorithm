```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from matplotlib.colors import ListedColormap
```

# Split Dataset

```python
x, y = make_classification(n_samples=500, n_features=2, n_classes=2, n_clusters_per_class=1, n_informative=2, n_redundant=0, random_state=42)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x
```

```python
y
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap='viridis', edgecolor='k', s=20)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Training Data Visualization')
plt.show()
```

# K-Nearest Neighbours algorithm

```python
k = 5
knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,average='micro')
recall = recall_score(y_test, y_pred,average='micro')
f1 = f1_score(y_test, y_pred,average='micro')
```

# the classifier's performance on the testing set

```python
df1=pd.DataFrame({'Accuracy':[accuracy],'Precision':[precision],'Recall':[recall],'F1 Score':[f1]})
df1
```

# Choose an appropriate distance metric and value for K

```python
k_values = range(1, 21)
cv_scores = []
```

```python
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(np.mean(scores))
```

```python
optimal_k = k_values[np.argmax(cv_scores)]
print(f"Optimal K: {optimal_k}")
```

```python
knn = KNeighborsClassifier(n_neighbors=optimal_k)
```

```python
knn.fit(x_train, y_train)
```

# Visualize the decision boundaries of the classifier

```python
h = .02
x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.5, cmap='viridis')
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, edgecolor='k', s=20, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title(f'KNN Decision Boundaries (Optimal K={optimal_k})')
plt.show()
```

# Interview Questions:
### 1. What are the key hyperparameters in KNN?

Key Hyperparameters in KNN:

*  Number of Neighbors (K): Determines how many nearest neighbors are used to make a prediction. A small K can lead to overfitting, while a large K can smooth out decision boundaries too much.
*  Distance Metric: Defines how distance is measured between points (e.g., Euclidean, Manhattan). The choice of metric can impact the performance and decision boundaries of the model.
*  Weights: Determines whether all neighbors contribute equally or if closer neighbors are given more weight (e.g., 'uniform' or 'distance').*

### **2. What distance metrics can be used in KNN?**

Distance Metrics in KNN:

*  Euclidean Distance: The most common metric, measures straight-line distance.
*  Manhattan Distance: Measures distance in grid-like paths, also known as L1 norm.
*  Minkowski Distance: Generalization of Euclidean and Manhattan, with a parameter p to define the type of distance.
Hamming Distance: Used for categorical variables, counts the number of different features.
*  Cosine Similarity: Measures the cosine of the angle between two vectors, used for text and high-dimensional data.
